## Task 3 :

1. Load the “tweets.cvs” file using pandas (Each line in the file represents a tweet)
2. Using nltk, tokenize the tweets.cvs into list of words.
3. It is often necessary to normalise the text for any NLP tasks. Unlike normal document or text, the
extracted tweets from Twitter need not be in a regular form.
    * Convert all the text in tweets.cvs to lower case letters
    * Clean tweets.cvs file by removing all the emoticons and slang words
    * Clean the tweets.cvs file by removing all the stop words
    * Punctuations are divided into priority. So punctuations like?, ., ! are retained and
        other like , “”, $, # etc are removed. 
    * Numbers are not important so we will remove numbers from tweets.cvs
    * Every tweet is followed by a URL that provides a link to the actual tweet. This is not
      required so remove url’s from tweets.csv file
    * save this corrected tweets in correctes_tweets.csv
    
4. Apply Potter Stemmer.
5. find and print the top 20 most frequent (unique words) 
from the tweets.cvs Plot the frequency distribution using pyPlot.

```python
import pandas as pd
df=pd.read_csv("tweets.csv",encoding="unicode_escape",header=None)
df
```

```python
list_tweets=list(df[1])
list_tweets
```

```python
from nltk.tokenize import word_tokenize
words=word_tokenize("".join(list_tweets))
words
```

```python
len(words)
```

```python
import re
from nltk.corpus import stopwords
ensw=stopwords.words("english")

def filter_url(word):
    regrex_pattern =re.compile(r"((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*")
    return regrex_pattern.sub(r'',word)


def filter_punc(word):
    regrex_pattern = re.compile(r"[#$%&'()\*\+-/:;<=>@[\]^_`{|}~]")
    return regrex_pattern.sub(r'',word)

def filter_digit(word):
    regrex_pattern =re.compile(r"[\d]+")
    return regrex_pattern.sub(r'',word)


def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)

def filter_stopword(word):
    if word not in ensw:
        return(word)
    else:
        return ""
 

def filterout(tweets):
    
    filter_tweet=[]
    for sent in tweets:
        sentence=[]
        for word in sent.split(" "):
            sentence.append((filter_digit(filter_stopword(deEmojify(filter_punc(filter_url(word.lower())))))))
        filter_tweet.append(" ".join(sentence))
    
    return filter_tweet

list_corrected_tweets=filterout(list_tweets)
list_corrected_tweets
```

```python
df=pd.DataFrame({"Correct Tweets": list_corrected_tweets})
df.to_csv("corrected_tweets.csv")
```

```python
from nltk.tokenize import word_tokenize
filter_words=word_tokenize("".join(list_corrected_tweets))
filter_words
```

```python
from nltk.stem import PorterStemmer
stem_words=[]
pstem=PorterStemmer()
for word in filter_words:
    stem_words.append(pstem.stem(word))
stem_words
```

```python
from nltk import FreqDist

fd=FreqDist(stem_words)
fd.most_common(20)
```

```python
fd.plot(20)
```

## Project Gender Classificaion

```python
from nltk.corpus import names
names.fileids()
```

```python
names.words('male.txt')
```

```python
from nltk.corpus import names
from nltk.classify import NaiveBayesClassifier

def gender(word):
    return {"last_letter ": word[-1]}

lab_names=[(name,"male") for name in names.words("male.txt")]+[(name,"female") for name in names.words("female.txt")]
train_data=[(gender(n),g)for (n,g) in lab_names]
model=NaiveBayesClassifier.train(train_data)

```

```python
model.classify(gender("Kiran"))
```
